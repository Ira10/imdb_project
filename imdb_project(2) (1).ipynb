```python
# Sort the dataframe with the 'profit' column as reference using the 'sort_values' function. 
#Make sure to set the argument 'ascending' to 'False'


df.sort_values(by = "profit", ascending=False, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.head(5)
```

```python
# Get the top 10 profitable movies by using position based indexing. Specify the rows till 10 (0-9)

top10=df.iloc[0:10]
top10
```

```python
plt.figure(figsize=(20,7))
sns.scatterplot(top10['budget'],top10['profit'])
```

```python
# Style / Theme setting
sns.set_style("whitegrid")

#Plot profit vs budget

plt.figure(figsize=[10,12])

plotProfitBudget=sns.jointplot(data=top10,x="budget", y='profit', height=8, ratio=8, color="Red", kind="reg")
plt.subplots_adjust(top=0.9)
plotProfitBudget.fig.suptitle('Scatter plot of Budget and Profit',fontsize= 18,fontweight= 7,color='Green')

plotProfitBudget.set_axis_labels('Budget', 'Profit', fontdict={'fontsize': 15,'fontweight': 6,'color': 'Green'})

plt.show()
```
